```python
import nsEVDx as ns
import numpy as np
from scipy.stats import genpareto
import matplotlib.pyplot as plt
import random
```

```python
np.random.seed(101)
data = genpareto.rvs(c=0.1, loc= 8, scale = 5,size = 48)
```

```python
plt.figure(figsize=(2,1.5),dpi=100)
plt.hist(data,density=True,bins=10,edgecolor='black')
plt.show()
```

#### Computing L-moments and L-moment ratios using ns.Lmoments()
Returns np.ndarray Array containing [n, mean, L1, L2, T3, T4], where

        - n: sample size
        - mean: sample mean
        - L1, L2: first and second L-moments
        - T3, T4: L-skewness and L-kurtosis

```python
ns.l_moments(data)
```

#### Fitting stationary GPD distribution using L-moments; ns.GPD_parsViaLM(data)

```python
shape, loc, scale = ns.GPD_parsViaLM(data)
print(f'Location: {loc}, Scale: {scale}, Shape: {shape}')
```

#### Fitting stationary GPD distribution using MLE; ns.EVD_parsViaMLE(data)

```python
shape, loc, scale = ns.EVD_parsViaMLE(data, genpareto)
print(f'Location: {loc}, Scale: {scale}, Shape: {shape}')
```
